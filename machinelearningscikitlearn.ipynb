```python
import pandas as pd
```

```python
heart_disease=pd.read_csv('heart.csv')
```

```python
heart_disease
```

```python
#create x
X=heart_disease.drop('target',axis=1)
X
Y=heart_disease['target']
Y
```

```python
from sklearn.ensemble import RandomForestClassifier
clf=RandomForestClassifier()
clf.get_params()
```

```python
#fit model
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3)
```

```python
clf.fit(X_train,Y_train)
```

```python
#Evaluate model
predicted_y=clf.predict(X_train)
```

```python
predicted_y
```

```python
clf.score(X_train,Y_train)
```

```python
clf.score(X_test,Y_test)
```

```python
#improve model
```

```python
for i in range(10,120,10):
    print(f'Runing Model with {i} estimators')
    clf=RandomForestClassifier(n_estimators=120).fit(X_train,Y_train)
    print(f'Accuracy is :{clf.score(X_test,Y_test)}')
```

```python
# save model
```

```python
import pickle
pickle.dump(clf,open('Heart Disease Predictor.pkl','wb'))
```

```python
load_model=pickle.load(open('Heart Disease Predictor.pkl','rb'))
load_model.score(X_test,Y_test)
```
